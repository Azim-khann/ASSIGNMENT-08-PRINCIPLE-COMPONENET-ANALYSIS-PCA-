#### Converting data to numpy array

```python
wine_array = wine_dup.values
wine_array
```

#### Normalizing the numerical data 

```python
from sklearn.preprocessing import scale
wine_norm = scale(wine_array)
wine_norm
```

#### PRINCIPLE COMPONENT ANALYSIS (PCA) IMPLEMETATION

APPLYING FIT TRANSFORM

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=13)
wine_pca = pca.fit_transform(wine_norm)
wine_pca
```

#### PCA Components matrix or covariance Matrix

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var_dup = np.cumsum(np.round(var,4)*100)
var_dup
```

```python
import matplotlib.pyplot as plt
plt.plot(var_dup, color='magenta')
```

#### FINAL DATAFRAME

```python
final_df = pd.concat([wine['Type'], pd.DataFrame(wine_pca[:,0:3], columns=['PC1','PC2','PC3'])], axis=1)
final_df
```

```python
import seaborn as sns
fig = plt.figure(figsize=(10, 5))
sns.scatterplot(data = final_df)
```

#### Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(20, 10))
dendrogram = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
Hir_clusters = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
Hir_clusters
```

```python
y = pd.DataFrame(Hir_clusters.fit_predict(wine_norm),columns = ['clustersid'])
y['clustersid'].value_counts()
```

#### Adding clusters to dataset

```python
wine_dup2 = wine.copy()
wine_dup2['clustersid']=Hir_clusters.labels_
wine_dup2
```

### K-Means Clustering

As we already have normalized data
Use Elbow Graph to find optimum number of  clusters (K value) from K values range
The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
random state can be anything from 0 to 42, but the same number to be used everytime,so that the results do

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Build Cluster algorithm using K=5

```python
clusters5 = KMeans(5,random_state=30).fit(wine_norm)
clusters5
```

```python
clusters5.labels_
```

```python
clusters3 = KMeans(5,random_state=30).fit(wine_norm)
clusters3
```
